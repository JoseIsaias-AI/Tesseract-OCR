```python
import pytesseract
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread('/kaggle/input/ocr1111/teste01.jpg')
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem);
plt.axis('off');
plt.show()
```

```python
texto = pytesseract.image_to_string(imagem)
print(texto)
```
